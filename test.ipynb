```python
def read_txt_n5(file_path):
    with open(file_path, "r") as file:
        for _ in range(5):
            line = file.readline()
            if not line:
                break  # In case the file has fewer than 5 lines
            print(line.strip())  # Print the line without trailing newline character
```

```python
read_txt_n5("Datasets/Amazon670K/trn_X_Xf.txt")
```

```python
read_txt_n5("Datasets/Amazon670K/trn_X_Y.txt")
```
